```python
import pandas as pd
from datetime import datetime
import altair as alt
```

```python
dados = pd.read_excel('20211013140246-precos-medios.xlsx')
dados
```

```python
ital_atac = dados.query('Produto == "Italiano - atacado"')
ital_atac = ital_atac.query('Região == "São Paulo (capital)"')
#ital_atac = ital_atac.query('Ano >= 2012 & Ano <= 2014')
#ital_atac

def transf_data(x):
  obj = str(int(x['Mês'])) + '-' + str(int(x['Ano']))
  date = datetime.strptime(obj, '%m-%Y')
  return date

ital_atac['ano_mes_data'] = ital_atac.apply(transf_data, axis = 1)
```

```python
alt.Chart(ital_atac).mark_line().encode(
    x = 'ano_mes_data:T',
    y = 'Preço:Q'
)
```
