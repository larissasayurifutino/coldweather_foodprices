Esse script descomprime os arquivos de IPCA em formato zip.

```python
import zipfile
import os
```

Subi os arquivos pelo menu lateral esquerdo do google colab.

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
 arr = os.listdir('ipca_zip')
 print(arr)
```

```python
for arq in arr:
  with zipfile.ZipFile('ipca_zip/' + arq, 'r') as zip_ref:
    zip_ref.extractall('ipca')
```
