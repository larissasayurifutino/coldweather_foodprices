```python
import pandas as pd
from plotnine import *
import altair as alt
```

```python
df = pd.read_csv('feijao.csv', sep = ";", decimal = ',')  

# Renomeando colunas
df = df.rename({'Safra 13/14' : '13/14',
 'Safra 14/15': '14/15',
 'Safra 15/16': '15/16',
 'Safra 16/17': '16/17',
 'Safra 17/18': '17/18',
 'Safra 18/19': '18/19',
 'Safra 19/20': '19/20',
 'Safra 20/21': '20/21'}, axis = 'columns')

df
```

```python
df_melt = pd.melt(df, 
        id_vars = 'reg',
        var_name = "Safra",
        value_name = "prod")
df_melt
```

```python

alt.Chart(df_melt).mark_line().encode(
    alt.X('Safra',
          title = 'Safra'),
    alt.Y('prod',
          title = 'Produção (em mil t)'),
    alt.Color('reg',
              title = 'Região')
).properties(width = 500, height = 250)
```
