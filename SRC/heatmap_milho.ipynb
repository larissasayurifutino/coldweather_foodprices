```python
import seaborn as sns
import numpy as np
import pandas as pd

df = pd.read_csv('milho_uf.csv', sep = ";", decimal = ',', index_col = 0)  
#df
```

```python
# Renomeando colunas
df = df.rename({'Safra 13/14' : '13/14',
 'Safra 14/15': '14/15',
 'Safra 15/16': '15/16',
 'Safra 16/17': '16/17',
 'Safra 17/18': '17/18',
 'Safra 18/19': '18/19',
 'Safra 19/20': '19/20',
 'Safra 20/21': '20/21'}, axis = 'columns')
```

```python
df = df.div(df.apply(max, axis = 1), axis = 0)*100
#df['UF'] = ['MT', 'MS', 'GO', 'DF', 'MG', 'SP', 'PR ']
df
```
